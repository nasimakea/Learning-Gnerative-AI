```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('german')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
senetences=nltk.sent_tokenize(paragraph)
```

```python
type(senetences)
```

```python
## Applt stop words , and filter and then aply stemming
```

```python
for i in range(len(senetences)):
    words=nltk.word_tokenize(senetences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    senetences[i]=' '.join(words)
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for i in range(len(senetences)):
    words=nltk.word_tokenize(senetences[i])
    words=[snowball_stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    senetences[i]=' '.join(words)
```

```python
senetences
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(senetences)):
    words=nltk.word_tokenize(senetences[i])
    words=[lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    senetences[i]=' '.join(words)
```
